```python
import json
import nltk
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import os
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns
register_matplotlib_converters()
```

```python
nltk.download('stopwords')
```

```python
speech_dir = "C:\\Users\katri\\Adv_prog_assignment3\\us_presidential_speeches\\"
files = os.listdir(speech_dir)

speeches = []
dates = []
for file in files:
    with open(speech_dir+file, "r") as infile:
        speech = json.load(infile)
        speeches.append(speech['Speech'])
        dates.append(speech['Date'])
```

```python
vectorizer = TfidfVectorizer(stop_words = stopwords.words('english'), ngram_range=(1,3))
X = vectorizer.fit_transform(speeches)
print(vectorizer.get_feature_names())
['and', 'document', 'first', 'is', 'one', 'second', 'the', 'third', 'this']
print(X.shape)
```

```python
term1 = 'america'
term2 = 'united states'
```

```python
term1_ind = vectorizer.get_feature_names().index(term1)
term2_ind = vectorizer.get_feature_names().index(term2)
```

```python
df = pd.DataFrame()
```

```python
df.head()
```

```python
dates2 = []
scores = []
terms = []
for i in range(1016):
    scores.append(X[i].T.toarray()[term1_ind][0])
    dates2.append(dates[i])
    terms.append(term1)
    scores.append(X[i].T.toarray()[term2_ind][0])
    dates2.append(dates[i])
    terms.append(term2)
```

```python
df['Date'] = dates2
df['Score'] = scores
df['Term'] = terms
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
nyc_chart = sns.lineplot(
    x="Date",
    y="Score",
    hue='Term',
    data=df
).set_title('Term by year')
plt.show()
```
